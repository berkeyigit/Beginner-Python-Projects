```python
class Kullanici:
    def __init__(self,isim,sifre):
        self.isim = isim
        self.__sifre = sifre

    def sifre_dogrula(self,girilen_sifre):
        return girilen_sifre == self.__sifre
    
class ATM:
    def __init__(self,kullanici):
        self.kullanici = Kullanici
        self.bakiye = 1000
        self.islemler = True

    def para_yatir(self):
        try:
            miktar = int(input("Lütfen yatirmak istediginiz meblayi giriniz : "))
            dogrula = input(f"{miktar} Yatiriyorsunuz. Emin misiniz ? (E/H)").upper()
            if dogrula =='E':
                if miktar <=0:
                    raise ValueError("Yatirilacak Miktar Pozitif Bir Sayi Olmalidir!")
                self.bakiye += miktar
                print(f"{miktar} TL hesabiniza yatirildi. Mevcut bakiyeniz : {self.bakiye}")
            else:
                print("İsleminiz İptal Ediliyor.")
                return ATM.menu_goster()
        except ValueError as hata:
            print(f"Hata: {hata}")

    def para_cek(self):
        try:
            miktar = int(input("Lütfen Cekmek İstediginiz Meblayi Giriniz : "))
            dogrula = input(f"{miktar} Cekiyorsunuz. Emin misiniz ? (E/H)").upper()
            if dogrula =='E':
                if miktar <=0:
                    raise ValueError("Cekmek İstediginiz Tutar Pozitif Bir Sayi Olmali")
                    
                if miktar > self.bakiye:
                    raise ValueError(f"Yeterli Bakiye yok. Bakiyeniz : {self.bakiye}")
                    
                if miktar % 10 != 0:
                    raise ValueError(f"Sadece '10' ve katlarini cekebilirsiniz.")
                    
                self.bakiye -= miktar
                print(f"{miktar} Hesabinizdan Cekildi. Mevcut Bakiyeniz : {self.bakiye}")
            else:
                print("İsleminiz İptal Ediliyor.")
                return ATM.menu_goster()
        except ValueError as hata:
            print(f"Hata: {hata}")

    def bakiye_sorgula(self):
        print(f"Bakiyeniz : {self.bakiye}")

    def para_gonder(self):
        try:
            kisi = input("Para Gondermek İstediginiz Kisinin Adini ve Soyadini Giriniz : ")
            miktar = int(input("Para Gondereceginiz Miktari Giriniz : "))
            dogrula = input(f"{kisi} kisisine {miktar} Gonderme Talebinizi Onaylayiniz. (E/H)").upper()
            if dogrula == 'E':
                if miktar <=0:
                    raise ValueError("Negatif Bir Miktar Gonderemezsiniz!")
                    
                if miktar > self.bakiye:
                    raise ValueError(f"Yeterli Bakiye yok. Bakiyeniz : {self.bakiye}")
                    
                self.bakiye -= miktar
                print(f"{miktar}, {kisi} Gonderildi. Mevcut Bakiyeniz : {self.bakiye}")
            else:
                print("İsleminiz İptal Ediliyor.")
                return ATM.menu_goster()
        except ValueError as hata:
            print(f"Hata: {hata}")

    def menu_goster(self):
        while self.islemler:
            print("~" * 50)
            print("1- Para Yatır\n2- Para Çek\n3- Bakiye Sorgula\n4- Para Gönder\n5- Çıkış Yap")
            print("~" * 50)

            try:
                islem = int(input("Lütfen Yapmak İstediginiz İslemi Giriniz : "))
                if islem == 1:
                    self.para_yatir()
                elif islem == 2:
                    self.para_cek()
                elif islem ==3:
                    self.bakiye_sorgula()
                elif islem ==4:
                    self.para_gonder()
                elif islem ==5:
                    print("Cikis yapiliyor...")
                    self.islemler = False
            except ValueError:
                print("Gcersiz Giris. Lutfen Bir Sayi Giriniz.")
    
Kullanici1 = Kullanici("Berke","2314")
girilen_sifre = input("Sifrenizi Girin: ")

if Kullanici1.sifre_dogrula(girilen_sifre):
    atm = ATM(Kullanici1)
    atm.menu_goster()
else:
    print("Hatali Sifre")

    

```
